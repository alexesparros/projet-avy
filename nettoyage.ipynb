```python
def clean_steam_data(df):
    # 1. Colonnes inutiles à supprimer
    columns_to_drop = [
        "background", "background_raw", "capsule_imagev5", "demos", "dlc",
        "legal_notice", "package_groups", "packages", "reviews",
        "screenshots", "support_info"
    ]
    df_clean = df.drop(columns=columns_to_drop, errors="ignore")
    
    # 2. Supprimer les doublons
    df_clean = df_clean.drop_duplicates()

    # 3. Supprimer les lignes sans nom de jeu
    if "name" in df_clean.columns:
        df_clean = df_clean[df_clean["name"].notna() & (df_clean["name"] != "")]

    # 4. Réinitialiser l’index avec "name" si elle existe
    if "name" in df_clean.columns:
        df_clean = df_clean.set_index("name")
    
    # 5. Sauvegarder en CSV
    df_clean.to_csv("df_clean.csv")

    return df_clean


```

```python
import pandas as pd
import ast
import re

def clean_steam_data(df):
    def safe_eval(x):
        try:
            return ast.literal_eval(x) if isinstance(x, str) else x
        except:
            return x

    def get_from_dict(x, key):
        x = safe_eval(x)
        return x.get(key) if isinstance(x, dict) else None

    def join_descriptions(x):
        x = safe_eval(x)
        return ", ".join(d.get("description", "") for d in x) if isinstance(x, list) else None

    def join_list(x):
        x = safe_eval(x)
        if isinstance(x, list):
            return ", ".join(str(i) for i in x)
        return str(x)

    def remove_html_tags(x):
        return re.sub(r"<.*?>", "", x) if isinstance(x, str) else x

    def extract_release_date(x):
        return get_from_dict(x, "date")

    def extract_min_text(x):
        if not isinstance(x, str):
            return None
        # Supprimer balises HTML s’il y en a
        x = re.sub(r"<.*?>", "", x)
        # Supprimer le préfixe "Minimum:" s’il est là
        x = x.strip()
        x = re.sub(r"^Minimum:\s*", "", x, flags=re.IGNORECASE)
        return "Minimum " + x if x else None

    # 1. Nettoyage de base
    columns_to_drop = ["background", "dlc", "packages", "coming_soon", "screenshots", "movies","about_the_game","short_description"]
    df = df.drop(columns=columns_to_drop, errors="ignore")
    df = df.drop_duplicates()
    df = df[df["name"].notna() & (df["name"] != "")]

    # 🔍 1.1 Filtrer uniquement les jeux
    if "type" in df.columns:
        df = df[df["type"] == "game"]

    # 2. Nettoyage des colonnes imbriquées
    if "achievements" in df.columns:
        df["achievements"] = df["achievements"].apply(lambda x: get_from_dict(x, "total"))
    if "categories" in df.columns:
        df["categories"] = df["categories"].apply(join_descriptions)
    if "genres" in df.columns:
        df["genres"] = df["genres"].apply(join_descriptions)
    if "content_descriptors" in df.columns:
        df["content_descriptors"] = df["content_descriptors"].apply(lambda x: get_from_dict(x, "notes"))
    if "developers" in df.columns:
        df["developers"] = df["developers"].apply(join_list)
    if "publishers" in df.columns:
        df["publishers"] = df["publishers"].apply(join_list)
    if "recommendations" in df.columns:
        df["recommendations"] = df["recommendations"].apply(lambda x: get_from_dict(x, "total"))
    if "release_date" in df.columns:
        df["release_date"] = df["release_date"].apply(extract_release_date)

    # 3. Extraction des prix
    if "price_overview" in df.columns:
        for key in ["currency", "discount_percent", "initial_formatted", "final_formatted"]:
            df[f"price_{key}"] = df["price_overview"].apply(lambda x: get_from_dict(x, key))
        df.drop(columns=["price_overview"], inplace=True)

    # 4. Nettoyage des langues
    if "supported_languages" in df.columns:
        df["supported_languages"] = df["supported_languages"].apply(remove_html_tags)

    # 5. Extraction des configurations minimales pour chaque OS
    for col in ["pc_requirements", "mac_requirements", "linux_requirements"]:
        if col in df.columns:
            df[col] = df[col].apply(lambda x: get_from_dict(x, "minimum"))
            df[col] = df[col].apply(extract_min_text)

    # 6. Réindexation et export
    df = df.set_index("name")
    df.to_csv("df_clean.csv")

    return df
```

```python
import re

def analyze_game_name_scripts(df):
    # Vérifie que l'index contient les noms de jeu
    if df.index.name != "name":
        print("⚠️ Attention : l'index ne contient pas les noms des jeux.")
        return None

    # Définir les regex pour différents scripts
    scripts = {
        "CJK (japonais/chinois/coréen)": r'[\u3040-\u30FF\u3400-\u4DBF\u4E00-\u9FFF\uF900-\uFAFF]',  # kana, kanji, hanzi
        "Coréen (Hangul)": r'[\uAC00-\uD7AF]',
        "Cyrillique": r'[\u0400-\u04FF]',
        "Arabe": r'[\u0600-\u06FF]',
        "Devanagari (hindi)": r'[\u0900-\u097F]',
        "Grec": r'[\u0370-\u03FF]',
        "Autres non-latins": r'[^\u0000-\u007F]',  # Tout ce qui n’est pas ASCII de base
    }

    total = len(df)
    results = {}

    # Analyse pour chaque script
    for label, pattern in scripts.items():
        regex = re.compile(pattern)
        matched = df[df.index.to_series().apply(lambda x: bool(regex.search(str(x))))]
        results[label] = {
            "count": len(matched),
            "percentage": len(matched) / total * 100,
            "sample": matched.index[:5].tolist()
        }

    # Affichage du résumé
    print("📊 Analyse des noms de jeux contenant des scripts non-latins :")
    for label, data in results.items():
        print(f"- {label} : {data['count']} jeux ({data['percentage']:.2f}%)")
        if data['count'] > 0:
            print(f"  Exemples : {data['sample']}")
    return results
```
